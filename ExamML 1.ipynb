# Model building

```python
#from the above all we can say the car_name and owner are not requied for the model
moddf=df.drop(['Car_Name','Owner'],axis=1)
```

```python
moddf
```

```python
x=moddf.drop('Selling_Price',axis=1)
```

```python
y=moddf['Selling_Price']
```

```python
#Since the data contains categorical data we use dummies for converting them to numercial data
x=pd.get_dummies(x)
x
```

```python
#Since the data has high variatio we use Standard Scaing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x.iloc[:,0:3]=sc.fit_transform(x.iloc[:,0:3])
x
```

```python
#for model building we use spliting the data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#Using Linear Legression model
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
model=lg.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
ypred
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR=RandomForestRegressor(n_estimators=100,random_state=1)
model1=RFR.fit(xtrain,ytrain)
```

```python
RFR_pred=model1.predict(xtest)
RFR_pred
```
